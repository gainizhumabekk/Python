```python
#Task 1
import random 
random.seed(105)
dice_rolls = [random.randint(1,7) for i in range(1000)]
```

```python
import matplotlib.pyplot as plt
plt.hist(dice_rolls, bins = [1,2,3,4,5,6,7], rwidth = 0.9, density = True)
plt.title("Distribution of Rolls")
plt.xlabel("Roll Values")
plt.ylabel("Probability")
plt.show()
```

```python
#Task 2
import numpy as np 
np.random.seed(105)
normal_data = np.random.normal(50, 15, 10000)

add_data = [random.randint(190, 211) for i in range(10)]
new_data = np.concatenate([normal_data, add_data])

mean_normal = np.mean(normal_data)
std_normal = np.std(normal_data)

mean_new = np.mean(new_data)
std_new = np.std(new_data)

print(f"Mean normal: {mean_normal}")
print(f"Mean new: {mean_new}")

print(f"Std normal: {std_normal}")
print(f"Std new: {std_new}")
```

```python
plt.subplot(1,2,1)
plt.hist(normal_data, bins = np.arange(10,220,10), rwidth = 0.9)
plt.title("Normal Distribution")

plt.subplot(1,2,2)
plt.hist(new_data, bins = np.arange(10,220,10), rwidth = 0.9)
plt.title("New Distribution")

plt.show()
#Разница не сильно заметная так как я добавила всего лишь 10 отклоняющихся значений по сравнению с 10000 оригинальными
```

```python
#Task 3
CityA = np.random.normal(50000, 15000, 1000)
CityB = np.random.normal(55000, 20000, 1000)
```

```python
import seaborn as sns
sns.kdeplot(CityA, label ="City A")
sns.kdeplot(CityB, label ="City B")

mean_a = np.mean(CityA)
mean_b = np.mean(CityB)
plt.axvline(mean_a, linestyle = "--")
plt.axvline(mean_b, linestyle = "--")

plt.title("Income Distributions of Cities A & B")
plt.xlabel("Income")
plt.ylabel("Density")
plt.legend()

plt.show()
```

```python
#Task 4 
serve_time = np.random.exponential(scale = 10, size = 1000)
sample = np.random.choice(serve_time, size = 100, replace = False)

plt.hist(sample, alpha = 0.5, bins = 20, rwidth = 0.9, density = True, label = "Histogram")
sns.kdeplot(sample, label = "KDE")

plt.title("Distribution of Serving Time (Sample)")
plt.xlabel("Serving Time")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(sample, line = "r")
plt.show()
```
