```python
#Task 1 
import matplotlib.pyplot as plt
plt.plot(["Mon","Tue","Wed","Thur","Fri","Sat","Sun"], [-10,-15, -13,-17,-16,-14,-10])
plt.title("Weather in Astana for the next week")
plt.xlabel("Days of Week")
plt.ylabel("Temperature, C")
plt.show()
```

```python
#Task 2 
plt.scatter([0,1, 2, 3, 4, 5, 6, 7, 8],[40,45,55,60,70,80,90,95,100])
plt.title("Relationship Between Study Hours and Grades")
plt.xlabel("Study Hours")
plt.ylabel("Grades in %")
plt.show()
```

```python
#Task 3
plt.bar(["Football", "Basketball", "Tennis", "Rugby", "Volleyball", "Swimming", "Skating"], [100,60,20,80,70,30,10])
plt.title("Number of Participants in Sports")
plt.xlabel("Type of Sports")
plt.ylabel("Number of Participants")
plt.show()
```

```python
#Task 4
import numpy as np 
plt.hist(np.random.randint(2, 6, 400),bins = 8)
plt.title("Distribution of Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.bar(["Espresso", "Latte", "Cappuccino"],[60, 80, 100])
plt.title("Coffee Sales per Day")
plt.xlabel("Type of Coffee")
plt.ylabel("Cups Sold")
plt.xticks(rotation=60)
plt.subplot(1,2,2)
plt.bar(["Croissants", "Baguettes", "Cookies", "Whoopies", "Macarons"],[40, 20, 60, 10, 50])
plt.title("Snack Sales per Day")
plt.xlabel("Type of Snack")
plt.ylabel("Items Sold")
plt.xticks(rotation=60)
plt.show()
```
