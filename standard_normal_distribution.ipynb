```python
#Task 1
import numpy as np
import random
import matplotlib.pyplot as plt

first = np.random.normal(0,1,1000)
second = np.random.normal(8,2,1000)

plt.hist(first, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Standard normal distr")
plt.hist(second, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Second normal distr")
plt.legend()
plt.show()
```

```python
second_shifted = second - np.mean(second)
plt.hist(first, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Standard normal distr")
plt.hist(second_shifted, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Second normal distr")
plt.legend()
plt.show()
```

```python
second_stretched = second_shifted / np.std(second)
plt.hist(first, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Standard normal distr")
plt.hist(second_stretched, rwidth = 0.9, bins = np.arange(-8,18,0.5), alpha = 0.5, label = "Second normal distr")
plt.legend()
plt.show()
```

```python
#Task 2
height = np.random.normal(170,10,1000)
height_shifted = height - np.mean(height)
height_stretched = height_shifted / np.std(height)

plt.subplot(1,2,1)
plt.hist(height, rwidth = 0.9, bins = 15)
plt.title("Original Height Distr")

plt.subplot(1,2,2)
plt.hist(height_stretched, rwidth = 0.9, bins = 15)
plt.title("Standard Normal Distr")

plt.show()
```

```python
#Task 3
from scipy.stats import norm

temperature = np.random.normal(22, 3, 1000)
high_temp = 27
low_temp = 15

probability_high = 1 - norm.cdf(high_temp, loc = np.mean(temperature), scale = np.std(temperature))
probability_low = norm.cdf(low_temp, loc = np.mean(temperature), scale = np.std(temperature))

print(f"High temperature percent: {probability_high*100:.2f}%")
print(f"Low temperature percent: {probability_low*100:.2f}%")

plt.hist(temperature, rwidth = 0.9, bins = 10, alpha = 0.5, label = "July Temperatures")
plt.hist(temperature[temperature>high_temp], bins = 5, color = "red", alpha = 0.5, label = "High Temps")
plt.hist(temperature[temperature<low_temp], bins = 5, color = "blue", alpha = 0.5, label = "Low Temps")

plt.title("July Temps Distr")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
