```python
import pandas as pd
from scipy.stats import mannwhitneyu, kstest, shapiro
import matplotlib.pyplot as plt
```

#### Задание 1
Вам подготовили датасет в файле compare_df.csv<br>
В нем есть всего две колонки sample1 и sample2<br>
Проведите Mann U Whitney тест чтобы проверить, идентично ли распределение двух этих сэмплов с уровнем значимости в 95%

```python
compare = pd.read_csv("compare_df.csv")
print(compare.head())
```

```python
u_statistic, p_value = mannwhitneyu(compare["sample1"], compare["sample2"], alternative = "two-sided")
print(u_statistic, p_value)

alpha = 0.05
if p_value < alpha:
    print("Есть разница в показателях")
else: 
    print("Существенной разницы в показателях нет")
```

#### Задание 2
Используя те же самые два сэмпла из compare_df, проведите Kolmogorov Smirnov тест для проверки идентичности распределений

```python
from scipy.stats import ks_2samp
ks_statistic, p_value = ks_2samp(compare["sample1"], compare["sample2"])
print(ks_statistic, p_value)

alpha = 0.05
if p_value < alpha:
    print("Есть разница в показателях")
else: 
    print("Существенной разницы в показателях нет")
```

#### Задание 3
Вам подготовили датасет в файле normality_test.csv<br>
В нем всего одна колонка sample<br>
Проведите Kolmogorov Smirnov тест, чтобы проверить нормальность данных с уровнем значимости в 95%

```python
one_sample = pd.read_csv("normality_test.csv")
print(one_sample.head())
```

```python
ks_statistic, p_value = kstest(one_sample["sample"], "norm")
print(ks_statistic, p_value)

alpha = 0.05
if p_value < alpha:
    print("Данные не распределены нормально")
else: 
    print("Данные распределены нормально")
```

#### Задание 4
Используя те же самые данные, проведите тест на нормальность используя Shapiro Wilk тест<br>
Какому тесту (ШапироУилк или КолмогоровСмирнов) вы доверяете больше и чей результат выберите?

```python
statistic, p_value = shapiro(one_sample["sample"])
print(statistic, p_value)

alpha = 0.05
if p_value < alpha:
    print("Данные не распределены нормально")
else: 
    print("Данные распределены нормально")
```

```python
# Shapiro–Wilk тест является более надежным для проверки нормальности
# В отличие от теста Колмогорова–Смирнова, он специально разработан для оценки нормальности распределения 
# и обладает большей статистической мощностью
```
