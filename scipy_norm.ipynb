```python
#Task 1
from scipy.stats import norm 
mean = 30
std_dev = 1
threshold = 29

probability = norm.cdf(threshold, loc = mean, scale = std_dev)
probability < 0.05

if probability < 0.05:
    print(f"Вероятность того, что порция меньше 29 мл, около {probability*100:.2f}%, что ниже допустимого уровня в 5%. Практика соответствует утверждениям кафе")
else:
    print(f"Вероятность того, что порция меньше 29 мл, около {probability*100:.2f}%, что выше допустимого уровня в 5%. Практика не соответствует утверждениям кафе")
```

```python
#Task 2
mean = 18
std_dev = 4
threshold = 22

probability = 1 - norm.cdf(threshold, loc = mean, scale = std_dev)
if probability < 0.01:
    print(f"Вероятность: {probability*100:.2f}%. Это не случайность, лекарство эффективно")
else:
    print(f"Снижение в > 22 мг/дл происходит примерно в {probability*100:.2f}% случаев даже у обычных препаратов. Это может быть случайность, не доказывает эффективности лекарства")
```

```python
#Task 3
import numpy as np
import random 
import matplotlib.pyplot as plt
import statsmodels.api as sm

random_data = (np.random.random(1000)*100).tolist() + np.random.exponential(10, 1000).tolist() + np.random.normal(50, 10, 1000).tolist()

mean_total = np.mean(random_data)
print(f"Mean of the original data: {mean_total}")

sample = np.random.choice(random_data, size = 300, replace = False)
mean_sample = np.mean(sample)
print(f"Mean of the sample: {mean_sample}")

plt.hist(random_data, rwidth = 0.9, density = True, bins = 20, alpha = 0.5)
plt.show()
sm.qqplot(np.array(random_data), line = "r")
plt.show()
```
