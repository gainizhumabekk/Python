## 1. Создайте колонку BMI (индекс массы тела), которая считается по формуле: BMI = (Вес в кг) / (Рост в метрах)^2. Учтите, что ваш текущий рост в см, не метры.

```python
data["BMI"] = data["Weight"] / (data["Height"]/100)**2
```

## 2. Посчитайте среднее значение, медиану, и моду для роста, веса, BMI.

```python
median_height = float(np.median(data["Height"]))
median_weight = float(np.median(data["Weight"]))
median_bmi = float(np.median(data["BMI"]))
print(f"Median height: {median_height}")
print(f"Median weight: {median_weight}")
print(f"Median BMI: {median_bmi}")
```

```python
import statistics 
mode_height = statistics.mode(data["Height"])
mode_weight = statistics.mode(data["Weight"])
mode_bmi = statistics.mode(data["BMI"])
print(f"Mode height: {mode_height}")
print(f"Mode weight: {mode_weight}")
print(f"Mode BMI: {mode_bmi}")
```

## 7. Посчитайте межквартильный размах для роста, веса, BMI.

```python
iqr_height = np.percentile(data["Height"], 75) - np.percentile(data["Height"], 25)
print(f"Interquartile range for height: {iqr_height}")
```

```python
iqr_weight = np.percentile(data["Weight"], 75) - np.percentile(data["Weight"], 25)
print(f"Interquartile range for weight: {iqr_weight}")
```

```python
iqr_bmi = np.percentile(data["BMI"], 75) - np.percentile(data["BMI"], 25)
print(f"Interquartile range for BMI: {iqr_bmi}")
```

## 8. Есть ли в данных какие-то выбросы для роста, веса, BMI? Как можно их найти? Используя межквартильный размах? Удалите все такие выбросы из таблицы.

```python
lower_height = np.percentile(data["Height"], 25) - 1.5*iqr_height
upper_height = np.percentile(data["Height"], 75) + 1.5*iqr_height

print(lower_height)
print(upper_height)
```

```python
lower_weight = np.percentile(data["Weight"], 25) - 1.5*iqr_weight
upper_weight = np.percentile(data["Weight"], 75) + 1.5*iqr_weight

print(lower_weight)
print(upper_weight)
```

```python
lower_bmi = np.percentile(data["BMI"], 25) - 1.5*iqr_bmi
upper_bmi = np.percentile(data["BMI"], 75) + 1.5*iqr_bmi

print(lower_bmi)
print(upper_bmi)
```

```python
cleaned_data = data[
    (data["Height"]>=lower_height) & (data["Height"]<=upper_height) &
    (data["Weight"]>=lower_weight) & (data["Weight"]<=upper_weight) &
    (data["BMI"]>=lower_bmi) & (data["BMI"]<=upper_bmi)]

cleaned_data.shape
```
