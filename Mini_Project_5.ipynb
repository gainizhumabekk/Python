```python
#Task 1
try:
    with open('users.txt', 'r') as users:
        while True:
            content = users.readline()
            if not content:
                break
            print(content)
            
except FileNotFoundError: 
    print("the file that you would like to read is not found")
    
except Exception: 
    print("some error occured")
```

```python
#Task 2

class Users: 
    def __init__(self, name, surname, mail, birth_year):
        self.name = name
        self.surname = surname
        self.mail = mail
        self.birth_year = int(birth_year)
        
    def getAge(self):
        return 2023 - self.birth_year

first_user = Users("Gaini", "Zhumabek", "gainizhumabek@gmail.com", 2001)
first_user.getAge()
```

```python
#Task 3
users_list = []

try:
    with open("users.txt", "r") as users:
        for i in users:
            splits = i.strip().split(",")
            
            if len(splits) == 5:
                user = Users(
                    splits[1],  # имя
                    splits[2],  # фамилия
                    splits[3],  # email
                    splits[4]   # год рождения
                        )
                users_list.append(user)

except FileNotFoundError:
    print("the file that you would like to read is not found")

except Exception:
    print("some error occured")

for user in users_list:
    print(user.name, user.getAge())
```

```python
#Task 4
def filter(users, low=30, high=50):
    for user in users:
        age = user.getAge()
        if low < age < high:
            print(f"{user.mail}, {age}")
filter(users_list)
```
