```python
import pandas as pd 
df = pd.read_csv("ДЗ_Pandas.csv")
```

```python
#Task 1
df.isnull().sum()
df.dropna(subset = ["Бюджет (млн $)"], inplace = True)
df.fillna({"Режиссер":"Неизвестен"}, inplace = True)
df.fillna({"Рейтинг": df["Рейтинг"].mean()}, inplace = True)
df.dropna(subset = ["Бюджет (млн $)", "Сборы (млн $)"], how = "all", inplace = True)
df
```

```python
#Task 2
#Группировка по Жанру
grouped_df = df.groupby("Жанр")
means = grouped_df[["Бюджет (млн $)", "Рейтинг"]].mean()
print(means)

#Группировка по Году Выпуска
group_year = df.groupby("Год выпуска")
total_sales = group_year["Сборы (млн $)"].agg(["sum", 'count'])
print(total_sales)

#Комбинированная Группировка
combined = df.groupby(["Жанр","Год выпуска"])
rating = combined["Рейтинг"].mean()
print(rating)
```

```python
#Task 3 Группировка и Сортировка
grouped_genre = df.groupby("Жанр")
grouped_genre["Рейтинг"].mean().reset_index().sort_values("Рейтинг", ascending = False)
```

```python
#Task 3 Анализ частоты
df["Год выпуска"].value_counts().sort_index().reset_index()
```

```python
#Task 3 Комбинированный Анализ
new_df = df.groupby(["Режиссер","Жанр"])
new_df["Сборы (млн $)"].sum().reset_index().sort_values("Сборы (млн $)", ascending = False)
```
